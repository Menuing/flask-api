```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('train.json')
df2 = pd.read_csv('epi_r.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.set_index('title', inplace=True)
df2.set_index('title', inplace=True)
```

```python
df2 = df2.loc[df.index]
df2.dropna(inplace=False)
```

```python
ingredients = pd.DataFrame()
ingredients['name'] = df2.columns.values[11:]
ingredients = ingredients[ingredients.index != 52]
ingredients = ingredients[ingredients.index != 156]
ingredients['name'] = [x.replace("'", ",") for x in ingredients.name]
```

```python
lowcost_col = []
fast_col = []
for name in df.index:
    lowcost = df2.loc[name]['3-ingredient recipes']
    if not isinstance(lowcost, np.float64):
        lowcost = lowcost.iloc[0]
    fast = df2.loc[name]['22-minute meals']
    if not isinstance(fast, np.float64):
        fast = fast.iloc[0]
        
    lowcost_col.append(lowcost)
    fast_col.append(fast)

df['3-ingredient recipes'] = lowcost_col
df['22-minute meals'] = fast_col
```

```python
df['ingredients'] = df.ingredients.map(lambda x: ';'.join(x))
df['ingredients'] = [x.replace("'", ",") for x in df.ingredients.values]
df['directions'] = df.directions.map(lambda x: ''.join(x))
df['directions'] = [x.replace("'", ".") for x in df.directions.values]
df['title'] = df.index.values
df['title'] = [x.replace("'", ".") for x in df.title.values]
df['3-ingredient recipes'] = df['3-ingredient recipes'] == 1
df['22-minute meals'] = df['22-minute meals'] == 1
df.set_index('title', inplace=True)
df = df[['calories', 'directions', 'fat', 'ingredients', 'protein', 'rating', 'sodium', '3-ingredient recipes', '22-minute meals']]
```
