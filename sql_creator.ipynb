```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('train.json')
df2 = pd.read_csv('epi_r.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.set_index('title', inplace=True)
df2.set_index('title', inplace=True)
```

```python
df2 = df2.loc[df.index]
df2.dropna(inplace=False)
```

```python
ingredients = pd.DataFrame()
ingredients['name'] = df2.columns.values[11:]
ingredients = ingredients[ingredients.index != 52]
ingredients = ingredients[ingredients.index != 156]
ingredients['name'] = [x.replace("'", ",") for x in ingredients.name]
```

```python
for i in ingredients.name:
    print("INSERT INTO ingredient (name) VALUES ('"+ i +"');", file=open("ingredient_queries.txt", "a", encoding="UTF-8"))
```

```python
df['ingredients'] = df.ingredients.map(lambda x: ';'.join(x))
df['ingredients'] = [x.replace("'", ",") for x in df.ingredients.values]
df['directions'] = df.directions.map(lambda x: ''.join(x))
df['directions'] = [x.replace("'", ".") for x in df.directions.values]
df['title'] = df.index.values
df['title'] = [x.replace("'", ".") for x in df.title.values]
df.set_index('title', inplace=True)
df = df[['calories', 'directions', 'fat', 'ingredients', 'protein', 'rating', 'sodium']]
```

```python
df
```

```python
for i in range(df.index.size):
    row = df.iloc[i]
    print("INSERT INTO recipe (name, instructions, proportions, calories, fat, protein, sodium, averagePuntuation, urlPhoto) \
          VALUES ('"+ row.name +"', '"+ row.directions +"', '"+ row.ingredients +"', "+ str(row.calories) +", "+ str(row.fat) +", \
         "+ str(row.protein) +", "+ str(row.sodium) +", "+ str(row.rating) +", '');", 
          file=open("recipe_queries.txt", "a", encoding="UTF-8"))
```

```python
df2['title'] = df2.index.values
df2['title'] = [x.replace("'", ".") for x in df2.title.values]
df2.set_index('title', inplace=True)
setting = set(df2.columns).intersection(set(ingredients.name))
rw = []
for i in setting:
    rw.append((i, df2.loc[df2[i] == 1].index.values))
```

```python
for i in range(len(rw)):
    for j in range(len(rw[i][1])):
        print("INSERT INTO recipe_ingredient (ingredient_id, recipe_id) (SELECT ing.id, re.id \
        FROM recipe re, ingredient ing WHERE re.name='"+rw[i][0]+"' and ing.name='"+rw[i][1][j]+"');", 
          file=open("recipe_ingredient.txt", "a", encoding="UTF-8"))
```
